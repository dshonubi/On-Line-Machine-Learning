```python
#Importing data
import numpy as np

M = np.loadtxt("M.txt")
v = np.loadtxt("v.txt").astype(int)
B = np.loadtxt("B.txt")
p = np.loadtxt("p.txt")
```

```python
def beta_dynamic(M,p,B,v):
    #beta goes from back to front so need to flip v so you start from the end
    v = np.flip(v)
    
    b_row = 0
    T_total = 0
    
    #initialiize beta array with row of 1s
    T_store = np.ones(len(M[0]))
    T_trac = 0
    T_storetemp = np.zeros(len(M[0]))
    
    for indxa, a in enumerate(v):
           
        if indxa == 0:
            
            T_storetemp = T_store.copy()
            
            for indxi, i in enumerate(M):
                for indxj, j in enumerate(i):
                    if indxi < len(M[0]):
                        
                        # Minus 1 as states are numbered from 0 and not 1 in my code
                        T_total = M[indxi][indxj] * T_storetemp[b_row] * B[b_row][v[indxa]-1] + T_total
                        b_row = b_row + 1

                        if b_row == len(M):
                            T_storetemp[T_trac] = T_total
                            T_total = 0
                            b_row = 0

                            if T_trac < len(M)-1:
                                T_trac = T_trac + 1
                            else:
                                T_trac = 0 
                                
            T_store = np.dstack((T_store,T_storetemp))                          
            
        elif (indxa > 0) and (indxa < 19):
            print(v[indxa])   
            T_temp = T_storetemp.copy()

            for indxi, i in enumerate(M):
                for indxj, j in enumerate(i):
                    if indxi < len(M[0]):

                        # Minus 1 as states are numbered from 0 and not 1 in my code
                        T_total = M[indxi][indxj] * T_temp[b_row] * B[b_row][v[indxa]-1] + T_total
                        b_row = b_row + 1
                        

                        if b_row == len(M[0]):
                            T_storetemp[T_trac] = T_total
                            T_total = 0
                            b_row = 0
                            
                            
                            if T_trac < len(M[0])-1:
                                T_trac = T_trac + 1
                            else:
                                T_trac = 0
             
            T_store = np.dstack((T_store,T_storetemp)) 
            
    T_store = np.flip(T_store)
    T_store = np.flipud(T_store[0])
    
    #prints beta 18 value 10 - minus 1 from both values as the code starts at 0
    #and NxT matrix means [9][17] is beta 18 value 10
    #print(T_store[9][17])
     
    return T_store

#testing
beta_dynamic(M,p,B,v)    
```
